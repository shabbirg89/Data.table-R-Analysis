```r
# Filter for year = 2018

df[year == 2018]
```

```r
#Filter for year = 2018 & color = gray
df[year == "2018" & color == "gray"]
```

```r
# Filter for cars with year = 2018 OR color = black
z <- df[year== "2018" | color == "Black"]

z
```

```r
#Filter for brand in kia ,dodge
z <- df[brand %in% c('kia','dodge')]

head(z)

# OR

#setkey(df, brand)

#z = df[c("dodge","kia")]
#head(z)
```

```r
# Select single coloum

#z <- df[,.(state)]
#head(z)

#OR in row manner
z <- df$state
head(z)
```

```r
# Select more than one columns

z <- df[,.(model,state)]
head(z,4)
```

```r
# Unique car production years

df[,unique(year)]
```

```r
# Maximum of the "mileage" column

df[,max(mileage)]
```

```r
# Apply functions on multiple columns

df[,.(Max_mileage = max(mileage),
      Min_mileage = min(mileage),
      Std_mileage = sd(mileage))]
```

```r
# Maximum mileage by year

df[, max(mileage),by = year]
```

```r
# Count by model and year

df[,.(Count = .N), 
   by = .(model, year)]
```

```r
# Call & Apply functions for grouping

df[,.(Count = .N), 
   by = .(Lessthan2005 = 
             year < 2005)]
```

```r
# Do the analysis on selected rows only

df[1:15,.(Max_mileage =
      max(mileage),
      Min_mileage =
      min(mileage),
      Std_mileage =
      sd(mileage))]
```

```r
# Remove a column

head(df[, brand := NULL])
```

```r
# Return the first occurance where title_status is clean vehicle

setkey(df, title_status)
df["clean vehicle", mult = "first"]
```

```r
# Return the last occurance where title_status is clean vehicle

setkey(df, title_status)
df["clean vehicle", mult = "last"]
```

```r
# Return a count for year in (2018 or 2015) and country usa

setkey(df, year, country)

df[.(c(2018,2015), "usa"), .N]
```

```r
# Return the splitted count for year in (2018 or 2015) and country usa
setkey(df, year, country)
df[.(c(2018,2015),  
      "usa"), .N,
   by =.EACHI]
```

```r
# Order in ascending order

tail(df[order(year)])
```

```r
# Order in descending order

tail(setorder(df, -year))
```

```r
# "year" in descending and "mileage" in ascending

tail(setorder(df, -year, mileage))
```

```r
# Rename columns
setnames(df,c("year","mileage"),c("YEAR", "MILEAGE"))
tail(df)
```
